### Import dependencies

```python
import os, json
from xml.dom import minidom
```

```python
svg_root = '../processed-svg/'
svg_paths = [path for path in os.listdir(svg_root) if path.endswith('.svg')]
```

```python
# Show first 5
svg_paths[:5]
```

### process each svg in root svg folder

```python
storage = {}

for svg_path in svg_paths:
    name = svg_path.replace('.svg', '').replace(' ','_').lower()
    doc = minidom.parse(os.path.join(svg_root,svg_path)) 
    view_box = doc.getElementsByTagName('svg')[0].getAttribute('viewBox')
   
    path_strings = [path.getAttribute('d') for path
                    in doc.getElementsByTagName('path')]
    data = {
        'viewBox': view_box,
        'pathStringList': path_strings
    }
    
    storage[name] = data
    doc.unlink()
```

```python
storage
```

### export to json with formatting 

```python
with open("./generated_svg_strings.json", "w") as outfile: 
    outfile.write(json.dumps(storage, indent = 2)) 
```
